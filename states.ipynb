```python
import csv

years = ["2017", "2018", "2019", "2020", "2021"]
states = ["Assam", "Kerala", "Goa"]
column_val = ["q1", "q2", "q3", "q4"]
a = int(input("Enter the size of the year dimensions: "))
b = int(input("Enter the size of the state dimensions: "))
c = int(input("Enter the size of the quarter attributes: "))

arr = [[[0 for _ in range(c)] for _ in range(b)] for _ in range(a)]

for i in range(a):
    for j in range(b):
        for k in range(c):
            print("Enter the quarter attributes", column_val[k], "of state:", states[j], "in year:", years[i])
            print("location: ", i, j, k)
            arr[i][j][k] = int(input())

with open("sample.csv", "w") as outdata:
    outdata.write("Data Representation\n")
    for i in range(a):
        for j in range(b):
            for k in range(c):
                print(arr[i][j][k], end="\t")
                outdata.write(str(arr[i][j][k]) + "\t")
            print("\t", end="")
            outdata.write("\t")
        print()
        outdata.write("\n")

    while True:
        print("Enter the dimension along which you want to slice:")
        print("1. year wise")
        print("2. state Row wise")
        print("3. quarter Attribute wise")
        print("4. Exit")
        option = int(input())

        if option == 1:
            print("Enter the year you want to extract")
            for i in range(a):
                print("year -", i+1, "-", years[i])
            year = int(input())
            outdata.write("years\n")
            for j in range(b):
                for k in range(c):
                    print(arr[year-1][j][k], end="\t")
                    outdata.write(str(arr[year-1][j][k]) + "\t")
                outdata.write("\t")
            outdata.write("\n")

        elif option == 2:
            print("Enter the state number you want to extract")
            for j in range(b):
                print("state -", j+1, "-", states[j])
            state = int(input())
            outdata.write("states\n")
            for i in range(a):
                for k in range(c):
                    print(arr[i][state-1][k], end="\t")
                    outdata.write(str(arr[i][state-1][k]) + "\t")
                outdata.write("\t")
            outdata.write("\n")

        elif option == 3:
            print("Enter the quarter Attribute you want to slice")
            for k in range(c):
                print(column_val[k], "-", k+1)
            att = int(input())
            outdata.write("quarter attributes\n")
            for i in range(a):
                for j in range(b):
                    print(arr[i][j][att-1], end="\t")
                    outdata.write(str(arr[i][j][att-1]) + "\t")
                outdata.write("\t")
            outdata.write("\n")
            outdata.close()

        elif option == 4:
            break
```

```python
import numpy as np

# get user input for array dimensions
serial_no = int(input("Enter the number of serial numbers: "))
no_of_years = int(input("Enter the number of years: "))
no_of_semesters = int(input("Enter the number of semesters: "))

# create 3D array with all values initialized to 0
arr = np.zeros((serial_no, no_of_years, no_of_semesters))

# get user input to populate the array
for i in range(serial_no):
    for j in range(no_of_years):
        for k in range(no_of_semesters):
            print(f"Enter value for array[{i}][{j}][{k}]:")
            arr[i][j][k] = int(input())

# perform slice and dice operations
while True:
    print("Select operation to perform:")
    print("1. Slice along the first dimension (serial no)")
    print("2. Slice along the second dimension (year)")
    print("3. Slice along the third dimension (semester)")
    print("4. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        serial_no_idx = int(input("Enter index for serial no dimension: "))
        print(arr[serial_no_idx])

    elif choice == 2:
        year_idx = int(input("Enter index for year dimension: "))
        print(arr[:, year_idx, :])

    elif choice == 3:
        semester_idx = int(input("Enter index for semester dimension: "))
        print(arr[:, :, semester_idx])

    elif choice == 4:
        break

    else:
        print("Invalid choice. Please try again.")
```
