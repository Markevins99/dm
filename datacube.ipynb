```python
import csv

# get user input for array dimensions
a = int(input("Enter the size of the semester dimensions: "))
b = int(input("Enter the size of the student dimensions: "))
c = int(input("Enter the size of the attribute dimensions: "))

arr = [[[0 for k in range(c)] for j in range(b)] for i in range(a)]

# get user input to populate the array
column_val = ["S.No", "Age", "Sleeping habit"]
for i in range(a):
    for j in range(b):
        for k in range(c):
            print(f"Enter the {column_val[k]} of student {j+1} in sem {i+1}")
            print(f"location\t{i}\t{j}\t{k}")
            arr[i][j][k] = int(input())

# perform slice, dice, and roll-up operations
while True:
    print("Select operation to perform:")
    print("1. Fetch attribute value")
    print("2. Slice the cube")
    print("3. Dice the cube")
    print("4. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        entered_S_No = int(input("Enter the student Serial Number: "))
        entered_semester = int(input("Enter the semester: "))
        print("Enter the attribute you want to fetch:")
        print("1. S.No\n2. Age\n3. Sleeping habit")
        chosen_attribute = int(input())

        if chosen_attribute < 1 or chosen_attribute > 3:
            print("Enter a valid attribute.")
        else:
            print(arr[entered_semester-1][entered_S_No-1][chosen_attribute-1])

            # write to a file
            with open('data_cube_sample.csv', mode='w', newline='') as file:
                writer = csv.writer(file)
                writer.writerow(['required attribute'])
                writer.writerow([arr[entered_semester-1][entered_S_No-1][chosen_attribute-1]])

    elif choice == 2:
        print("Select operation to perform:")
        print("1. Slice by semester")
        print("2. Slice by student")
        print("3. Exit")
        slice_choice = int(input("Enter choice: "))

        if slice_choice == 1:
            entered_semester = int(input("Enter the semester: "))
            slice_result = [arr[entered_semester-1][j] for j in range(b)]
        elif slice_choice == 2:
            entered_S_No = int(input("Enter the student Serial Number: "))
            slice_result = [[arr[i][entered_S_No-1][j] for j in range(c)] for i in range(a)]
        elif slice_choice == 3:
            break
        else:
            print("Invalid choice. Please try again.")
            continue

        # print and write to file
        print("Slice result:")
        for i in range(len(slice_result)):
            print(f"Semester {entered_semester}, Student {i+1}: {slice_result[i]}")
            with open('data_cube_sample.csv', mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([f"Semester {entered_semester}, Student {i+1}"])
                writer.writerow(slice_result[i])
        print("Slice result written to file.")

    elif choice == 3:
        print("Select operation to perform:")
        print("1. Dice by semester")
        print("2. Dice by student")
        print("3. Exit")
        dice_choice = int(input("Enter choice: "))

        if dice_choice == 1:
            entered_semester = int(input("Enter the semester: "))
            dice_result = arr[entered_semester-1]
        elif dice_choice == 2:
            entered_S_No = int(input("Enter the student Serial Number: "))
            dice_result = [[arr[i][entered_S_No-1][j] for j in range(c)] for i in range(a)]
        elif dice_choice == 3:
            break
        else:
            print("Invalid choice. Please try again.")
            continue

        # print and write to file
        print("Dice result:")
        for i in range(len(dice_result)):
            for j in range(len(dice_result[i])):
                print(f"Semester {i+1}, Student {j+1}: {dice_result[i][j]}")
                with open('data_cube_sample.csv', mode='a', newline='') as file:
                    writer = csv.writer(file)
                    writer.writerow([f"Semester {i+1}, Student {j+1}"])
                    writer.writerow(dice_result[i][j])
        print("Dice result written to file.")

    elif choice == 4:
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
import csv

# get user input for array dimensions
a = int(input("Enter the size of the semester dimensions: "))
b = int(input("Enter the size of the student dimensions: "))
c = int(input("Enter the size of the attribute dimensions: "))

arr = [[[0 for k in range(c)] for j in range(b)] for i in range(a)]

# get user input to populate the array
column_val = ["S.No", "Age", "Sleeping habit"]
for i in range(a):
    for j in range(b):
        for k in range(c):
            print(f"Enter the {column_val[k]} of student {j+1} in sem {i+1}")
            print(f"location\t{i}\t{j}\t{k}")
            arr[i][j][k] = int(input())

# perform slice and dice operations on the array
print("Performing slice and dice operations on the array...")

# slice the array to get the attributes of all students in the first semester
sem1_attrs = []
for j in range(b):
    sem1_attrs.append([arr[0][j][k] for k in range(c)])

# dice the array to get the sleeping habits of all students in all semesters
sleeping_habits = []
for i in range(a):
    sem_sleep_habits = []
    for j in range(b):
        sem_sleep_habits.append(arr[i][j][2])
    sleeping_habits.append(sem_sleep_habits)

# save the results in a CSV file
print("Saving the results in a CSV file...")

with open('slice_and_dice.csv', mode='w') as file:
    writer = csv.writer(file)
    writer.writerow(column_val)
    writer.writerow(["Semester 1 Attributes"])
    writer.writerows(sem1_attrs)
    for i in range(a):
        writer.writerow([f"Semester {i+1} Sleeping Habits"])
        writer.writerows([[sleeping_habits[i][j]] for j in range(b)])
```
